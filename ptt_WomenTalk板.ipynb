```python
#https://github.com/chrisyang-tw/Simple_PTT_Crawler
import requests
from bs4 import BeautifulSoup
import urllib.parse
import csv
import jieba.analyse
import jieba
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np

has_kotex = ''
has_whisper = ''
has_sofy = ''
has_kao = ''
has_carnation= ''

jieba.set_dictionary("dict.txt.big")
jieba.load_userdict('userdict.txt')

#stopword list:
stopword_list=[]
file = open('stopword_chinese.txt','r',encoding="utf-8")  #watch out the encoding!
Lines = file.readlines() 
for line in Lines: 
    line = line.replace('\n', '')
    stopword_list.append(line)
file.flush()
file.close()
stopwords = {}.fromkeys(stopword_list)
    
index = str(input('想抓取哪個ptt看板？(ex: Movie版請輸入 https://www.ptt.cc/bbs/movie/index.html)：\n'))
pages = eval(input('想抓取幾頁呢？ex: 5：'))

not_exist = BeautifulSoup('<a>(本文已被刪除)</a>', 'lxml').a ## '本文已被刪除'的結構不同，自行生成<a>


def get_pages(num): ## 要爬幾頁
    page_url = index
    all_articles = []

    for j in range(num):
        articles, next_link = get_articles_on_ptt(page_url)
        all_articles += articles
        page_url = urllib.parse.urljoin(index, next_link) ## 將上一頁按鈕的網址和 index 網址比對後取代
    
    return all_articles

def get_articles_on_ptt(url): ## 爬取一頁的文章
    global has_kotex
    global has_whisper
    global has_sofy
    global has_kao
    global has_carnation

    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml') ## 得到網頁原始碼
    articles = [] 
    
    for i in soup.find_all('div', 'r-ent'):
        comments = ""
        #main2 = ''
        meta = i.find('div', 'title').find('a') or not_exist
        response2 = requests.get('https://www.ptt.cc'+ i.find('div', 'title').find('a').get('href'))
        soup2 = BeautifulSoup(response2.text, 'lxml')
        main = soup2.find('div', 'bbs-screen').getText()
        for c in soup2.find_all('div', 'push'):
            main = main.replace(c.getText(), '')
            comments = comments + c.getText()
        for pc in soup2.find_all('span', 'push-content'):
            if('靠得住' in pc.getText() or 'Kotex' in pc.getText()):
                has_kotex = has_kotex + pc.getText()
            if('好自在' in c.getText()):
                has_whisper = has_whisper + pc.getText()
            if('蘇菲' in pc.getText() or '蘇X' in pc.getText()):
                has_sofy = has_sofy + pc.getText()
            if('蕾妮亞' in pc.getText()):
                has_kao = has_kao + pc.getText()
            if('康乃馨' in pc.getText()):
                has_carnation = has_carnation + pc.getText()
        #print(comments)   
        main = main.replace(soup2.find('div', 'article-metaline-right').getText(), '')
        main = main.replace('--', '')
        for r in soup2.find_all('span', 'f2'):
            main = main.replace(r.getText(), '')
        for m in soup2.find_all('div', 'article-metaline'):
            main = main.replace(m.getText(), '')
        #print(main)
#         for pc in soup2.find_all('span', 'push-content'):
#             main2 = main + pc.getText()
#         if('靠得住' in main2 or 'Kotex' in main2):
#             has_kotex = has_kotex + main2
#         if('好自在' in main2):
#             has_whisper = has_whisper + main2
#         if('蘇菲' in main2):
#             has_sofy = has_sofy + main2
#         if('蕾妮亞' in main2):
#             has_kao = has_kao + main2
#         if('康乃馨' in main2):
#             has_carnation = has_carnation + main2
        articles.append({
            'title': meta.getText().strip(), ## strip 去除頭尾字符，預設是空白
            'push': i.find('div', 'nrec').getText(),
            'date': i.find('div', 'date').getText(),
            'author': i.find('div', 'author').getText(),
            'link': 'https://www.ptt.cc'+ i.find('div', 'title').find('a').get('href'),
            'content': main,
            'comments': comments
        })

    next_link = soup.find('div', 'btn-group-paging').find_all('a', 'btn')[1].get('href') ## 控制頁面選項(上一頁)
    return articles, next_link

data = get_pages(pages)

for k in data: ## 輸出至螢幕
    print(k['push'], k['title'], k['date'], k['author'], k['link'])
    
csv_or_not = input('輸入 y 以匯出成csv檔，輸入其他結束程式：')

#https://blog.gtwang.org/programming/python-csv-file-reading-and-writing-tutorial/
if csv_or_not == 'y':
    board = index.split('/')[-2] ## 取出看板名
    with open('ptt_WomenTalk衛生棉.csv', 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['推文數', '標題', '發文日期', '作者ID', '連結', '內文', '留言'])
        for l in data:
            writer.writerow([l['push'], l['title'], l['date'], l['author'] , l['link'], l['content'], l['comments']])
    csvfile.close()
    print('csv檔案已儲存在您的資料夾中。')
else:
    quit()

def generate_wordcloud(keywords, stopwords, file_path):
    wc = WordCloud(font_path = 'msyh.ttf', background_color = "white",
                    max_words = 2000, stopwords = stopwords)
    wc.generate_from_frequencies(keywords)
    plt.imshow(wc)
    plt.axis("off")
    plt.figure(figsize = (10, 6), dpi = 100)
    plt.show()
    wc.to_file(file_path)
    
def get_keywords(article, topN):
    keywords = {}
    jieba.analyse.set_stop_words('stopword_chinese.txt')
    tags = jieba.analyse.extract_tags(article, topK = topN, withWeight = True)
    for tag, weight in tags:
        keywords[tag] = weight
    return keywords

print(has_kotex)
print(has_whisper)
print(has_sofy)

if(has_kotex):
    keywords = get_keywords(has_kotex, 50)
    generate_wordcloud(keywords, stopwords, "ptt靠得住關鍵字wordcloud.jpg")
if(has_whisper):
    keywords = get_keywords(has_whisper, 50)
    generate_wordcloud(keywords, stopwords, "ptt好自在關鍵字wordcloud.jpg")
if(has_sofy):
    keywords = get_keywords(has_sofy, 50)
    generate_wordcloud(keywords, stopwords, "ptt蘇菲關鍵字wordcloud.jpg")
if(has_kao):
    keywords = get_keywords(has_kao, 50)
    generate_wordcloud(keywords, stopwords, "ptt蕾妮雅關鍵字wordcloud.jpg")
if(has_carnation):
    keywords = get_keywords(has_carnation, 50)
    generate_wordcloud(keywords, stopwords, "ptt康乃馨關鍵字wordcloud.jpg")
```
